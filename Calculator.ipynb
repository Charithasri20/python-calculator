# Simple Calculator — Python Project

## What This Does
A command-line calculator that performs addition,
subtraction, multiplication, and division.

## Features
- Supports +  -  *  / operations
- Handles division by zero error
- Handles invalid number input
- Loops until user types quit

## Concepts Used
- Functions with parameters and return values
- While loops
- try/except error handling
- if/elif/else conditions

## How to Run
Run the code cell below

```python
# Simple Calculator — Day 3-4 Python Project
# B.Tech AIML — 30 Day Learning Challenge

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero!"
    return a / b

def calculator():
    print("=" * 40)
    print("      SIMPLE CALCULATOR")
    print("=" * 40)
    print("Operations: +  -  *  /")
    print("Type 'quit' to exit")
    print("=" * 40)

    while True:
        num1_input = input("\nEnter first number: ")
        if num1_input.lower() == "quit":
            print("Thank you for using the calculator!")
            break

        num2_input = input("Enter second number: ")
        if num2_input.lower() == "quit":
            print("Thank you for using the calculator!")
            break

        operator = input("Enter operator (+, -, *, /): ")

        try:
            num1 = float(num1_input)
            num2 = float(num2_input)
        except ValueError:
            print("Invalid input! Please enter valid numbers.")
            continue

        if operator == "+":
            result = add(num1, num2)
            print(f"Result: {num1} + {num2} = {result}")

        elif operator == "-":
            result = subtract(num1, num2)
            print(f"Result: {num1} - {num2} = {result}")

        elif operator == "*":
            result = multiply(num1, num2)
            print(f"Result: {num1} * {num2} = {result}")

        elif operator == "/":
            result = divide(num1, num2)
            print(f"Result: {num1} / {num2} = {result}")

        else:
            print("Invalid operator! Please use +, -, *, /")

# Run the calculator
calculator()
```
